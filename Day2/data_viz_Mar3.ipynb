```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.array([10, 13, 19, 27, 33, 39, 41])
y = np.array([15, 18, 24, 33, 37, 42, 48])
z = np.array([50, 44, 40, 35, 29, 22, 15])
```

```python
# line plot
```

```python
plt.title("trend analysis")
plt.xlabel("x-values")
plt.ylabel("y-values")

plt.plot(x, y, marker = 'o', mfc = 'red', color = 'purple',
        ls= '--');
plt.grid()
# markers -> *, o, <, ^, >, H
```

```python
df = pd.read_csv('student.csv')

tips = sns.load_dataset('tips')
```

```python
df;
```

```python
plt.title("student marks")
plt.xlabel("names")
plt.ylabel("marks")
plt.xticks(rotation = 45)

plt.bar(df['name'], df['marks'], color = 'black', 
               edgecolor = 'yellow');

plt.grid()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
iris['species'] = le.fit_transform(iris['species'])
```

```python
plt.figure(figsize=(12, 9))
plt.scatter(iris['sepal_width'], iris['sepal_length'],
           c = iris['species']);
```

```python
sns.scatterplot(x = 'sepal_width', y ='sepal_length',
               data = iris, hue = 'species');
```
