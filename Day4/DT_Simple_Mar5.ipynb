```python
sns.scatterplot(x = 'Budget', y = 'Celebrities', data = df,
               hue = 'Class');
```

```python
x = df[['Budget', 'Celebrities']]
y = df['Class']
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state= 0)

dt.fit(x, y)
```

```python
budget = 16000000
cele = 6
```

```python
dt.predict([[budget, cele]])
```
